#### About Dataset:
This dataset gives information about the daily air quality index in Allegheny County. AQI is a numerical value that determines how safe the air is in a given location for the people there. AQI also has a description that goes along with certain ranges of values. For example, AQI values from 51-100 would be considered moderate.

#### Why We Chose This Dataset:
We chose this dataset because it is helpful for determining the healthiest neighborhood in Pittsburgh based on environmental factors, and analyzing the air that people breathe is very important to make a valid conclusion. If people are breathing air that has a high AQI value, then those people are in more danger for a health concern than people breathing air with a low AQI value. This helps us determine what areas are healthy, which makes the dataset very useful.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv("AQI.csv")
air_quality.head(10)
```

```python
air_quality.info()
```

#### What we did so far:
1. We imported the needed libraries
2. We loaded the dataset
3. We checked the top 10 values in the dataset to make sure everything is loaded properly
4. We checked the datatype of each category for later use

```python
good_sites = []
duplicate = 0
count = 0
for i in air_quality["index_value"]:
    if i <= 50:
        if air_quality["site"].iloc[count] in good_sites:
            duplicate += 1
        else:
            good_sites.append(air_quality["site"].iloc[count])
    count += 1
print(air_quality["description"].unique())
print()
print("Location of sites with low AQI:", good_sites)
print()
print("All Unique Sites:", air_quality["site"].unique())
```

#### What we learned:
The way AQI works is that if the value is less than or equal to 50, then that location has good air quality
1. There are 5 different categories for AQI in this dataset
2. There are a total of 15 site categories in this dataset
3. The locations with a 2 at the end exist because that means there are different locations with the same name
4. All 15 of the sites have a point in time where it had good air quality
5. We cannot eliminate any location from having the best air quality as of now

```python
average = air_quality.groupby('site')['index_value'].mean()
print(average)
```

```python
list_sites = list(average.keys())
average_values = list(average.values)
plt.figure(figsize=(30,15))
plt.bar(list_sites, average_values)
plt.xticks(fontsize=20, rotation=45)
plt.yticks(fontsize=20)
plt.ylabel("AQI Values", fontsize = 30)
plt.xlabel("Sites", fontsize = 30)
plt.title("Average AQI for Each Location", fontsize = 40)
plt.show()
```

#### What we accomplished:
1. We found the average AQI value for each of the locations
2. We displayed the values on a graph for easier interpretation
3. We learnt that Lawrenceville 2, West Mifflin, Flag Plaza, Glassport High Street, and Liberty are the 5 sites with the best average AQI

```python
good_count = air_quality[air_quality["description"] == "Good"].groupby("site").size()
total_count = air_quality.groupby("site").size() - good_count
plt.figure(figsize=(30,15))
plt.bar(list_sites, good_count, color='r')
plt.bar(list_sites, total_count, bottom=good_count, color='b')
plt.xticks(fontsize=20, rotation=45)
plt.yticks(fontsize=20)
plt.ylabel("Occurrences", fontsize = 30)
plt.xlabel("Sites", fontsize = 30)
plt.title("Good vs Total Description Occurrences For Each Location", fontsize = 40)
plt.show()
```

```python
ratio = good_count / total_count
print(ratio)
```

#### What we accomplished:
1. We found the ratio of good descriptions over total descriptions to see which locations have the most good descriptions
2. We graphed our results on a stacked bar chart to visualize the results
3. We used a table to verify our graph and ensure correct conclusions
4. We learnt that West Mifflin, Harrison Township, Lawrenceville 2, Flag Plaza, and Glassport High Street are the 5 sites with the best ratios
